#### Importing necessary libraries

```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

#### Loading our dataset!

```python
df = pd.read_csv('cancer_classification.csv')
df.head()
```

```python
df.describe().transpose()
```

#### Performing some EDA

```python
sns.countplot(x='benign_0__mal_1', data=df)
```

```python
df.corr()['benign_0__mal_1'][:-1].sort_values().plot(kind='bar')
```

```python
plt.figure(figsize=(12, 12))
sns.heatmap(df.corr(), annot=False)
```

```python
X = df.drop('benign_0__mal_1', axis=1).values           # predictors
y = df['benign_0__mal_1'].values            # target
```

#### Splitting our data into training and testing subsets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

#### Scaling our data to get optimum results

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

#### Building our model!

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential([Dense(units=30, activation='relu'),
                    Dense(units=15, activation='relu'),
                    Dense(units=1, activation='sigmoid')])          # binary classification!
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
losses.head()
```

```python
losses.plot()           # example of overfitting!
```

Redefining our model by enabling `EarlyStopping` to prevent overfitting

```python
model = Sequential([Dense(units=30, activation='relu'),
                    Dense(units=15, activation='relu'),
                    Dense(units=1, activation='sigmoid')])          # binary classification!
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test),
          callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(data=model.history.history)
model_loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(units=30, activation='relu'))
model.add(Dropout(rate=0.5))

model.add(Dense(units=15, activation='relu'))
model.add(Dropout(rate=0.5))

# binary classification!
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test),
          callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
predictions = (model.predict(X_test) > 0.5).astype(int)
predictions[:5]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
