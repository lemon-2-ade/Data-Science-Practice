```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set_style('darkgrid')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
```

### Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cancer['target'], cmap='bwr')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

In `pca.components_`, each row represents a principal components and each column actually relates back to it's original feature!

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
df_comp
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df_comp, cmap='plasma', annot=False)
```
