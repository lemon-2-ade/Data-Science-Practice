```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
df_feat = pd.DataFrame(data=cancer['data'], columns=cancer['feature_names'])
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = df_feat, cancer.target
# y = cancer['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(gamma='auto')
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions), end="\n")
print(classification_report(y_test, predictions))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C' : [0.1, 1, 10, 100, 1000],
    'gamma' : [1, 0.1, 0.01, 0.001, 0.0001]
}
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, verbose=2)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test, grid_predictions), end="\n")
print(classification_report(y_test, grid_predictions))
```
