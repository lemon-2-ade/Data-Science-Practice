```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set_style('darkgrid')
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=4,
                  cluster_std=1.43, random_state=101)
```

```python
plt.scatter(x=X[:, 0], y=X[:, 1], c=y, cmap='viridis')
# plt.grid(visible=True)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10, 6))

ax1.set_title("K Means")
ax1.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')

ax2.set_title("Original")
ax2.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
```
