```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
sin_values = np.sin(x)
cos_values = np.cos(x)
x_tan = np.linspace(-np.pi/4, np.pi/4, 256, endpoint=True)
tan_values = np.tan(x_tan)
```

```python
fig, axes = plt.subplots(figsize=(8, 4))
# axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, sin_values, color='red', lw=2, alpha=0.7)          # plotting the sin curve
axes.plot(x, cos_values, color='green', lw=2, alpha=0.7)        # plotting the cosine curve
axes.plot(x_tan, tan_values, color='purple', lw=2, alpha=1)
axes.set_xlabel('Angle in Radians'); axes.set_ylabel('Trigonometric Plots')
axes.set_title('Trigonometric Plot')
```
