```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

#### Functional Method

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

#### Creating 2 Plots on the same Canvas

```python
plt.subplot(1, 2, 1)            # plt.subplot(rows, columns, plot_no.) <- syntax
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

#### Object - Oriented Method

```python
# object oriented method
fig = plt.figure()              # creating a figure object
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])               # fig.add_axes([left, bottom, width, height]) <- syntax

axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()          # creating the figure object

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)          # 'axes' contains a list of matplotib objects which can be iterated through

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')

plt.tight_layout()          # takes care ofany overlapping of the plots
```
